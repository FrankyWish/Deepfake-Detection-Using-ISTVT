```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import timm
import numpy as np
import random
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from collections import defaultdict
import argparse
import glob
import shutil


torch.manual_seed(42)
np.random.seed(42)
random.seed(42)


class FlatDeepFakeDataset(Dataset):
    def __init__(self, root_dir, frames_per_video=30, transform=None, max_clips=None):
        
        self.samples = []
        self.transform = transform or self.default_transforms()
        self.frames_per_video = frames_per_video
        
        real_dir = os.path.join(root_dir, 'real')
        fake_dir = os.path.join(root_dir, 'fake')
        
        real_groups = self._group_files(real_dir, is_real=True)
        fake_groups = self._group_files(fake_dir, is_real=False)
        
        self._create_clips(real_groups, real_dir, label=0)
        self._create_clips(fake_groups, fake_dir, label=1)
        
        self.balance_dataset()
        
        if max_clips:
            self.samples = self.samples[:max_clips]
        
        print(f"Loaded {len(self.samples)} video clips from {root_dir}")
        print(f"Real clips: {sum(1 for s in self.samples if s['label'] == 0)}")
        print(f"Fake clips: {sum(1 for s in self.samples if s['label'] == 1)}")
    
    def _group_files(self, directory, is_real):
        groups = defaultdict(list)
        for filename in os.listdir(directory):
            if not filename.endswith('.png'):
                continue
            
            name_parts = filename.split('.')[0].split('_')
            
            if is_real:
                if len(name_parts) < 2:
                    continue
                try:
                    video_id = name_parts[0]
                    frame_num = int(name_parts[-1])
                except ValueError:
                    continue
            else:
                if len(name_parts) < 3:
                    continue
                try:
                    video_id = f"{name_parts[0]}_{name_parts[1]}"
                    frame_num = int(name_parts[-1])
                except ValueError:
                    continue
            
            groups[video_id].append((frame_num, filename))
        
        return groups

    def _create_clips(self, groups, base_dir, label):
        for video_id, frames in groups.items():
            frames.sort(key=lambda x: x[0])
            frame_paths = [os.path.join(base_dir, fn) for _, fn in frames]
            for start_idx in range(0, len(frame_paths), self.frames_per_video):
                end_idx = start_idx + self.frames_per_video
                if end_idx <= len(frame_paths):
                    clip = {
                        'paths': frame_paths[start_idx:end_idx],
                        'label': label,
                        'video_id': video_id
                    }
                    self.samples.append(clip)

    def balance_dataset(self):
        real_clips = [s for s in self.samples if s['label'] == 0]
        fake_clips = [s for s in self.samples if s['label'] == 1]
        
        if not real_clips or not fake_clips:
            print("Warning: Missing real or fake samples. No balancing performed.")
            return
        
        min_count = min(len(real_clips), len(fake_clips))
        real_clips = random.sample(real_clips, min_count)
        fake_clips = random.sample(fake_clips, min_count)
        
        self.samples = real_clips + fake_clips
        random.shuffle(self.samples)
        
        print(f"Balanced dataset: {min_count} real clips, {min_count} fake clips")

    def default_transforms(self):
        return transforms.Compose([
            transforms.Resize((224, 224)), 
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                                 std=[0.229, 0.224, 0.225])
        ])

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        sample = self.samples[idx]
        frames = []
        for path in sample['paths']:
            try:
                img = Image.open(path).convert('RGB')
                if self.transform:
                    img = self.transform(img)
                frames.append(img)
            except Exception as e:
                print(f"Error loading {path}: {e}")
                # Create blank image as fallback
                blank = Image.new('RGB', (224, 224), (0, 0, 0))
                frames.append(self.transform(blank))
                
        return torch.stack(frames), sample['label']


class MultiHeadAttention(nn.Module):
    def __init__(self, dim, num_heads=8, qkv_bias=False, attn_drop=0., proj_drop=0.):
        super().__init__()
        self.num_heads = num_heads
        head_dim = dim // num_heads
        self.scale = head_dim ** -0.5

        self.qkv = nn.Linear(dim, dim * 3, bias=qkv_bias)
        self.attn_drop = nn.Dropout(attn_drop)
        self.proj = nn.Linear(dim, dim)
        self.proj_drop = nn.Dropout(proj_drop)
        
        self.attention_maps = None

    def forward(self, x):
        B, N, C = x.shape
        qkv = self.qkv(x).reshape(B, N, 3, self.num_heads, C // self.num_heads).permute(2, 0, 3, 1, 4)
        q, k, v = qkv[0], qkv[1], qkv[2]

        attn = (q @ k.transpose(-2, -1)) * self.scale
        attn = attn.softmax(dim=-1)
        attn = self.attn_drop(attn)
        
        self.attention_maps = attn.detach().cpu()
        
        x = (attn @ v).transpose(1, 2).reshape(B, N, C)
        x = self.proj(x)
        x = self.proj_drop(x)
        return x


class TransformerBlock(nn.Module):
    def __init__(self, dim, num_heads, mlp_ratio=4., qkv_bias=False, drop=0., attn_drop=0.):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.attn = MultiHeadAttention(
            dim, 
            num_heads=num_heads, 
            qkv_bias=qkv_bias, 
            attn_drop=attn_drop, 
            proj_drop=drop
        )
        self.norm2 = nn.LayerNorm(dim)
        mlp_hidden_dim = int(dim * mlp_ratio)
        self.mlp = nn.Sequential(
            nn.Linear(dim, mlp_hidden_dim),
            nn.GELU(),
            nn.Dropout(drop),
            nn.Linear(mlp_hidden_dim, dim),
            nn.Dropout(drop)
        )

    def forward(self, x):
        x = x + self.attn(self.norm1(x))
        x = x + self.mlp(self.norm2(x))
        return x


class SpatialTransformer(nn.Module):
    def __init__(self, backbone='vit_base_patch16_224', pretrained=True):
        super().__init__()
        self.vit = timm.create_model(
            backbone, 
            pretrained=pretrained, 
            num_classes=0,  
            features_only=False
        )
        self.embed_dim = self.vit.embed_dim
        
    def forward(self, x):
        B, T, C, H, W = x.shape
        x = x.view(B * T, C, H, W)
        
        features = self.vit.forward_features(x)
        
        cls_tokens = features[:, 0]
        
        return cls_tokens.view(B, T, -1)


class TemporalTransformer(nn.Module):
    def __init__(self, dim, depth, num_heads, mlp_ratio=4., qkv_bias=True, drop_rate=0., attn_drop_rate=0.):
        super().__init__()
        self.cls_token = nn.Parameter(torch.zeros(1, 1, dim))
        self.pos_embed = nn.Parameter(torch.zeros(1, 101, dim))  
        self.blocks = nn.ModuleList([
            TransformerBlock(
                dim=dim, 
                num_heads=num_heads, 
                mlp_ratio=mlp_ratio, 
                qkv_bias=qkv_bias, 
                drop=drop_rate, 
                attn_drop=attn_drop_rate
            )
            for _ in range(depth)
        ])
        self.norm = nn.LayerNorm(dim)
        
        nn.init.trunc_normal_(self.cls_token, std=.02)
        nn.init.trunc_normal_(self.pos_embed, std=.02)
        
    def forward(self, x):
        B, T, D = x.shape
        
        cls_tokens = self.cls_token.expand(B, -1, -1)
        x = torch.cat((cls_tokens, x), dim=1)
        
        pos_embed = self.pos_embed[:, :T+1]
        x = x + pos_embed
        
        for blk in self.blocks:
            x = blk(x)
            
        x = self.norm(x)
        
        return x[:, 0]


class ISTVT(nn.Module):
    def __init__(self, 
                 spatial_backbone='vit_base_patch16_224',
                 spatial_pretrained=True,
                 temporal_depth=2,
                 temporal_num_heads=8,
                 temporal_mlp_ratio=4.,
                 num_classes=2,
                 qkv_bias=True,
                 drop_rate=0.1,
                 attn_drop_rate=0.1):
        super().__init__()
        
        self.spatial_transformer = SpatialTransformer(
            backbone=spatial_backbone,
            pretrained=spatial_pretrained
        )
        
        self.temporal_transformer = TemporalTransformer(
            dim=self.spatial_transformer.embed_dim,
            depth=temporal_depth,
            num_heads=temporal_num_heads,
            mlp_ratio=temporal_mlp_ratio,
            qkv_bias=qkv_bias,
            drop_rate=drop_rate,
            attn_drop_rate=attn_drop_rate
        )
        
        self.head = nn.Sequential(
            nn.LayerNorm(self.spatial_transformer.embed_dim),
            nn.Linear(self.spatial_transformer.embed_dim, num_classes)
        )
        
    def forward(self, x):
        spatial_features = self.spatial_transformer(x)    
        temporal_features = self.temporal_transformer(spatial_features)        
        return self.head(temporal_features)
    
    def get_attention_maps(self):
        attn_maps = []
        for blk in self.temporal_transformer.blocks:
            attn_maps.append(blk.attn.attention_maps)
        return attn_maps


def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    
    pbar = tqdm(dataloader, desc="Training")
    for inputs, labels in pbar:
        inputs, labels = inputs.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item() * inputs.size(0)
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()
        
        pbar.set_postfix({
            'loss': loss.item(),
            'acc': f'{100.*correct/total:.1f}%'
        })
    
    avg_loss = total_loss / total
    accuracy = 100. * correct / total
    return avg_loss, accuracy


def validate(model, dataloader, criterion, device, set_name="Validation"):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    all_labels = []
    all_preds = []
    
    with torch.no_grad():
        pbar = tqdm(dataloader, desc=f"{set_name}")
        for inputs, labels in pbar:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            total_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
            
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())
            
            pbar.set_postfix({
                'acc': f'{100.*correct/total:.1f}%'
            })
    
    avg_loss = total_loss / total
    accuracy = 100. * correct / total
    
    print(f"\n{set_name} Classification Report:")
    print(classification_report(
        all_labels, 
        all_preds, 
        target_names=["Real", "Fake"],
        digits=4
    ))
    
    return avg_loss, accuracy


def visualize_attention(attention_maps, frame_indices=None, save_path=None):
    
    if frame_indices is None:
        frame_indices = list(range(10))  
    
    num_heads = attention_maps[0].shape[1]
    num_blocks = len(attention_maps)
    
    plt.figure(figsize=(15, 5 * num_blocks))
    
    for block_idx, attn in enumerate(attention_maps):
        cls_attn = attn[:, :, 0, 1:]          
        avg_attn = cls_attn.mean(0)[:, frame_indices] 
        
        for head_idx in range(num_heads):
            plt.subplot(num_blocks, num_heads, block_idx * num_heads + head_idx + 1)
            plt.bar(range(len(frame_indices)), avg_attn[head_idx])
            plt.title(f"Block {block_idx+1} Head {head_idx+1}")
            plt.xlabel("Frame Index")
            plt.ylabel("Attention Weight")
            plt.xticks(range(len(frame_indices)), frame_indices)
    
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300)
        print(f"Saved attention visualization to {save_path}")
    plt.show()


def visualize_on_test_set(model, test_loader, device):
    print("\nVisualizing attention on test samples...")
    model.eval()
    
    inputs, labels = next(iter(test_loader))
    inputs = inputs.to(device)
    with torch.no_grad():
        outputs = model(inputs)
    
    attention_maps = model.get_attention_maps()
    sample_idx = 0
    sample_input = inputs[sample_idx].cpu()
    sample_label = labels[sample_idx].item()
    visualize_attention(
        attention_maps, 
        frame_indices=range(min(10, inputs.shape[1])),
        save_path=f'test_attention_{"real" if sample_label == 0 else "fake"}.png'
    )
    
    plt.figure(figsize=(15, 5))
    for i in range(min(5, inputs.shape[1])):
        plt.subplot(1, 5, i+1)
        img = sample_input[i].permute(1, 2, 0).numpy()
        img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
        img = np.clip(img, 0, 1)
        plt.imshow(img)
        plt.title(f"Frame {i}")
        plt.axis('off')
    plt.suptitle(f"Sample Test Clip ({'Real' if sample_label == 0 else 'Fake'})")
    plt.savefig('test_sample_frames.png', dpi=300)
    plt.show()


def predict_video(model, video_path, frames=30, device='cuda'):
    os.makedirs("temp_frames", exist_ok=True)
    
    os.system(f"ffmpeg -i {video_path} -vf fps=30 temp_frames/%04d.png")
    
    frame_files = sorted(glob.glob("temp_frames/*.png"))[:frames]
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    
    frames = []
    for f in frame_files:
        try:
            img = Image.open(f).convert('RGB')
            frames.append(transform(img))
        except:
            blank = Image.new('RGB', (224, 224), (0, 0, 0))
            frames.append(transform(blank))
    
    clip = torch.stack(frames).unsqueeze(0).to(device)
    
    model.eval()
    with torch.no_grad():
        output = model(clip)
        prob = torch.softmax(output, 1)[0,1].item()
    
    shutil.rmtree("temp_frames")
    
    return prob

def main(args):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    print("\n===== Creating Datasets =====")
    train_dataset = FlatDeepFakeDataset(
        root_dir=os.path.join(args.data_root, 'train'),
        frames_per_video=args.frames_per_video,
        max_clips=args.max_train_clips
    )
    val_dataset = FlatDeepFakeDataset(
        root_dir=os.path.join(args.data_root, 'validation'),
        frames_per_video=args.frames_per_video,
        max_clips=args.max_val_clips
    )
    test_dataset = FlatDeepFakeDataset(
        root_dir=os.path.join(args.data_root, 'test'),
        frames_per_video=args.frames_per_video,
        max_clips=args.max_test_clips
    )
    train_loader = DataLoader(
        train_dataset,
        batch_size=args.batch_size,
        shuffle=True,
        num_workers=args.num_workers,
        pin_memory=True
    )
    val_loader = DataLoader(
        val_dataset,
        batch_size=args.batch_size,
        shuffle=False,
        num_workers=args.num_workers,
        pin_memory=True
    )
    test_loader = DataLoader(
        test_dataset,
        batch_size=args.batch_size,
        shuffle=False,
        num_workers=args.num_workers,
        pin_memory=True
    )
    
    print(f"Train clips: {len(train_dataset)}, Validation clips: {len(val_dataset)}, Test clips: {len(test_dataset)}")
    
    print("\n===== Creating ISTVT Model =====")
    model = ISTVT(
        spatial_backbone=args.spatial_backbone,
        temporal_depth=args.temporal_depth,
        temporal_num_heads=args.temporal_heads,
        drop_rate=args.drop_rate
    ).to(device)
    
    print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
    
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(
        model.parameters(),
        lr=args.learning_rate,
        weight_decay=args.weight_decay
    )
    
    scheduler = optim.lr_scheduler.CosineAnnealingLR(
        optimizer, 
        T_max=args.epochs,
        eta_min=args.min_lr
    )
    
    history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': []
    }
    
    best_val_acc = 0.0
    
    print("\n===== Starting Training =====")
    for epoch in range(args.epochs):
        print(f"\nEpoch {epoch+1}/{args.epochs}")
        
        train_loss, train_acc = train_epoch(
            model, train_loader, criterion, optimizer, device
        )
        
        val_loss, val_acc = validate(
            model, val_loader, criterion, device, "Validation"
        )
        
        scheduler.step()
        
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'best_model.pth')
            print(f"Saved new best model with val accuracy: {val_acc:.2f}%")
        
        print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
              f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")
    
    torch.save(model.state_dict(), "final_model.pth")
    print("\nSaved final model to final_model.pth")
    
    plt.figure(figsize=(12, 10))
    
    plt.subplot(2, 1, 1)
    plt.plot(history['train_loss'], 'b-', label='Train')
    plt.plot(history['val_loss'], 'r-', label='Validation')
    plt.title('Loss Curve')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 1, 2)
    plt.plot(history['train_acc'], 'b-', label='Train')
    plt.plot(history['val_acc'], 'r-', label='Validation')
    plt.title('Accuracy Curve')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('training_history.png', dpi=300)
    print("Saved training history plot to training_history.png")
    
    print("\n===== Evaluating on Test Set =====")
    model.load_state_dict(torch.load('best_model.pth'))
    test_loss, test_acc = validate(
        model, test_loader, criterion, device, "Test"
    )
    print(f"Final Test Accuracy: {test_acc:.2f}%")
    
    print("\n===== Visualizing Attention on Test Set =====")
    visualize_on_test_set(model, test_loader, device)
    
    if args.sample_video:
        print("\n===== Running Sample Inference =====")
        fake_prob = predict_video(model, args.sample_video, args.frames_per_video, device)
        print(f"Predicted fake probability: {fake_prob:.2%}")


class Args:
    def __init__(self):
        self.data_root = "final_FFpp" 
        self.frames_per_video = 15
        self.max_train_clips = 100
        self.max_val_clips = 20
        self.max_test_clips = 20
        self.batch_size = 2            
        self.num_workers = 0            
        self.epochs = 5
        self.learning_rate = 1e-4
        self.min_lr = 1e-6
        self.weight_decay = 0.01
        self.drop_rate = 0.1
        self.spatial_backbone = 'vit_small_patch16_224'
        self.temporal_depth = 2
        self.temporal_heads = 4
        self.sample_video = None

args = Args()


def main_jupyter(args):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    print("\n===== Creating Datasets =====")
    train_dataset = FlatDeepFakeDataset(
        root_dir=os.path.join(args.data_root, 'train'),
        frames_per_video=args.frames_per_video,
        max_clips=args.max_train_clips
    )
    
    val_dataset = FlatDeepFakeDataset(
        root_dir=os.path.join(args.data_root, 'validation'),
        frames_per_video=args.frames_per_video,
        max_clips=args.max_val_clips
    )
    
    test_dataset = FlatDeepFakeDataset(
        root_dir=os.path.join(args.data_root, 'test'),
        frames_per_video=args.frames_per_video,
        max_clips=args.max_test_clips
    )
    
    train_loader = DataLoader(
        train_dataset,
        batch_size=args.batch_size,
        shuffle=True,
        num_workers=0,  
        pin_memory=True
    )
    
    val_loader = DataLoader(
        val_dataset,
        batch_size=args.batch_size,
        shuffle=False,
        num_workers=0,  
        pin_memory=True
    )
    
    test_loader = DataLoader(
        test_dataset,
        batch_size=args.batch_size,
        shuffle=False,
        num_workers=0,  
        pin_memory=True
    )
    
    print(f"Train clips: {len(train_dataset)}, Validation clips: {len(val_dataset)}, Test clips: {len(test_dataset)}")
    
    print("\n===== Creating ISTVT Model =====")
    model = ISTVT(
        spatial_backbone=args.spatial_backbone,
        temporal_depth=args.temporal_depth,
        temporal_num_heads=args.temporal_heads,
        drop_rate=args.drop_rate
    ).to(device)
    
    print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
    
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(
        model.parameters(),
        lr=args.learning_rate,
        weight_decay=args.weight_decay
    )
    
    scheduler = optim.lr_scheduler.CosineAnnealingLR(
        optimizer, 
        T_max=args.epochs,
        eta_min=args.min_lr
    )
    
    history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': []
    }
    
    best_val_acc = 0.0
    
    print("\n===== Starting Training =====")
    for epoch in range(args.epochs):
        print(f"\nEpoch {epoch+1}/{args.epochs}")
        
        train_loss, train_acc = train_epoch(
            model, train_loader, criterion, optimizer, device
        )
        
        val_loss, val_acc = validate(
            model, val_loader, criterion, device, "Validation"
        )
        
        scheduler.step()
        
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'best_model.pth')
            print(f"Saved new best model with val accuracy: {val_acc:.2f}%")
        
        print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
              f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")
    
    torch.save(model.state_dict(), "final_model.pth")
    print("\nSaved final model to final_model.pth")
    
    plt.figure(figsize=(12, 10))
    
    plt.subplot(2, 1, 1)
    plt.plot(history['train_loss'], 'b-', label='Train')
    plt.plot(history['val_loss'], 'r-', label='Validation')
    plt.title('Loss Curve')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 1, 2)
    plt.plot(history['train_acc'], 'b-', label='Train')
    plt.plot(history['val_acc'], 'r-', label='Validation')
    plt.title('Accuracy Curve')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('training_history.png', dpi=300)
    plt.show()
    print("Saved training history plot to training_history.png")
    
    print("\n===== Evaluating on Test Set =====")
    model.load_state_dict(torch.load('best_model.pth'))
    test_loss, test_acc = validate(
        model, test_loader, criterion, device, "Test"
    )
    print(f"Final Test Accuracy: {test_acc:.2f}%")
    
    print("\n===== Visualizing Attention on Test Set =====")
    visualize_on_test_set(model, test_loader, device)
    
    if args.sample_video:
        print("\n===== Running Sample Inference =====")
        fake_prob = predict_video(model, args.sample_video, args.frames_per_video, device)
        print(f"Predicted fake probability: {fake_prob:.2%}")
    
    return model


class Args:
    def __init__(self):
        self.data_root = "final_FFpp"  
        self.frames_per_video = 15
        self.max_train_clips = 100
        self.max_val_clips = 20
        self.max_test_clips = 20
        self.batch_size = 2            
        self.num_workers = 0            
        self.epochs = 5
        self.learning_rate = 1e-4
        self.min_lr = 1e-6
        self.weight_decay = 0.01
        self.drop_rate = 0.1
        self.spatial_backbone = 'vit_small_patch16_224'
        self.temporal_depth = 2
        self.temporal_heads = 4
        self.sample_video = None

args = Args()

print("Starting training in Jupyter notebook...")
trained_model = main_jupyter(args)
print("Training completed!")
```
