```python
import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import timm
import numpy as np
from tqdm import tqdm
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

torch.manual_seed(42)
np.random.seed(42)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

class DeepFakeDataset(Dataset):
    def __init__(self, root_dir, frames_per_video=5, transform=None, max_samples=None):
        self.samples = []
        self.transform = transform or transforms.Compose([
            transforms.Resize((299, 299)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
        ])

        real_dir = os.path.join(root_dir, 'real')
        fake_dir = os.path.join(root_dir, 'fake')

        real_frames = {}
        for fname in os.listdir(real_dir):
            if fname.endswith('.png'):
                parts = fname.split('_')
                video_id = parts[0]
                frame_idx = int(parts[1].split('.')[0])
                if video_id not in real_frames:
                    real_frames[video_id] = []
                real_frames[video_id].append((frame_idx, os.path.join(real_dir, fname)))

        fake_frames = {}
        for fname in os.listdir(fake_dir):
            if fname.endswith('.png'):
                parts = fname.split('_')
                video_id = parts[0] + '_' + parts[1]  
                frame_idx = int(parts[2].split('.')[0])
                if video_id not in fake_frames:
                    fake_frames[video_id] = []
                fake_frames[video_id].append((frame_idx, os.path.join(fake_dir, fname)))

        for video_id, frames in real_frames.items():
            frames = sorted(frames, key=lambda x: x[0])
            frame_paths = [path for _, path in frames]
            for i in range(0, len(frame_paths), frames_per_video):
                clip_paths = frame_paths[i:i+frames_per_video]
                if len(clip_paths) == frames_per_video:
                    self.samples.append((clip_paths, 0))

        for video_id, frames in fake_frames.items():
            frames = sorted(frames, key=lambda x: x[0])
            frame_paths = [path for _, path in frames]
            for i in range(0, len(frame_paths), frames_per_video):
                clip_paths = frame_paths[i:i+frames_per_video]
                if len(clip_paths) == frames_per_video:
                    self.samples.append((clip_paths, 1))

        if max_samples:
            self.samples = self.samples[:max_samples]

        print(f"Loaded {len(self.samples)} video clips from {root_dir}")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        frame_paths, label = self.samples[idx]
        frames = []
        for path in frame_paths:
            img = Image.open(path).convert('RGB')
            if self.transform:
                img = self.transform(img)
            frames.append(img)
        return torch.stack(frames), torch.tensor(label, dtype=torch.long)


class DeepFakeDetector(nn.Module):
    def __init__(self, num_frames=5):
        super().__init__()
        self.backbone = timm.create_model("xception", pretrained=True, num_classes=0)

        self.lstm = nn.LSTM(
            input_size=2048,
            hidden_size=512,
            num_layers=1,
            batch_first=True
        )

        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, 2)
        )  

    def forward(self, x):
        batch_size, num_frames, C, H, W = x.shape

        features = []
        for t in range(num_frames):
            frame_features = self.backbone(x[:, t])
            features.append(frame_features)

        features = torch.stack(features, dim=1)

        lstm_out, _ = self.lstm(features)

        last_out = lstm_out[:, -1, :]

        return self.classifier(last_out)


def train_model(model, train_loader, val_loader, epochs=10, lr=0.0001):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=2, factor=0.5)

    best_acc = 0.0
    history = {'train_loss': [], 'val_acc': []}

    for epoch in range(epochs):
        # Training
        model.train()
        running_loss = 0.0
        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")

        for inputs, labels in progress_bar:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            progress_bar.set_postfix(loss=loss.item())

        epoch_loss = running_loss / len(train_loader.dataset)
        history['train_loss'].append(epoch_loss)

        val_acc = evaluate_model(model, val_loader)
        history['val_acc'].append(val_acc)
        scheduler.step(epoch_loss)  

        print(f"Epoch {epoch+1}/{epochs} - Loss: {epoch_loss:.4f}, Val Acc: {val_acc:.4f}")

        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), 'best_model.pth')
            print(f"Saved new best model with accuracy: {val_acc:.4f}")

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Training Loss')
    plt.title('Training Loss')
    plt.xlabel('Epochs')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['val_acc'], label='Validation Accuracy')
    plt.title('Validation Accuracy')
    plt.xlabel('Epochs')
    plt.legend()
    plt.savefig('training_history.png')
    plt.show()

    return model


def evaluate_model(model, loader):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in tqdm(loader, desc="Evaluating"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    acc = accuracy_score(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, target_names=["Real", "Fake"])
    print(f"Accuracy: {acc:.4f}")
    print("Classification Report:")
    print(report)

    return acc


if __name__ == "__main__":
    data_config = {
        'frames_per_video': 5,
        'batch_size': 8
    }

    train_dataset = DeepFakeDataset(
        root_dir="/content/drive/MyDrive/Facefp/train",
        frames_per_video=data_config['frames_per_video'],
        max_samples=200  
    )

    val_dataset = DeepFakeDataset(
        root_dir="/content/drive/MyDrive/Facefp/valid",
        frames_per_video=data_config['frames_per_video'],
        max_samples=50
    )

    train_loader = DataLoader(
        train_dataset,
        batch_size=data_config['batch_size'],
        shuffle=True,
        num_workers=2
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=data_config['batch_size'],
        shuffle=False,
        num_workers=2
    )

    print(f"Train samples: {len(train_dataset)}, Validation samples: {len(val_dataset)}")

    model = DeepFakeDetector(num_frames=data_config['frames_per_video'])
    print(f"Model created with {sum(p.numel() for p in model.parameters()):,} parameters")

    trained_model = train_model(
        model,
        train_loader,
        val_loader,
        epochs=15,
        lr=0.0001
    )

    print("\nFinal Evaluation on Validation Set:")
    evaluate_model(trained_model, val_loader)

    test_dataset = DeepFakeDataset(
        root_dir="/content/drive/MyDrive/Facefp/test",
        frames_per_video=data_config['frames_per_video']
    )
    test_loader = DataLoader(test_dataset, batch_size=data_config['batch_size'], shuffle=False)

    print("\nFinal Evaluation on Test Set:")
    evaluate_model(trained_model, test_loader)

    torch.save(trained_model.state_dict(), "final_model.pth")
    print("Model saved as final_model.pth")
```
